## üìä Scoring Methodology

Our composite scoring system evaluates repositories across three weighted dimensions:

### üåü Popularity Score (40% weight)
**Measures community engagement and recognition**
- Stars: Community appreciation
- Forks: Code reuse and collaboration
- Watchers: Ongoing interest
- *Normalized to 0-100 scale*

### ‚ö° Activity Score (35% weight)
**Measures development velocity and engagement**
- Commits: Activity over 30d, 90d, 365d windows
- Contributors: Team size and collaboration
- Recency: Time since last update
- *Higher scores indicate active development*

### üè• Health Score (25% weight)
**Measures repository maturity and professionalism**
- README: Documentation presence (99.3% coverage)
- License: Legal clarity (94.2% coverage)
- Contributing guidelines: Onboarding support
- Code of conduct: Community standards
- *Indicates professional project management*

### üéØ Overall Score Calculation

```
Overall Score = (Popularity √ó 0.40) + (Activity √ó 0.35) + (Health √ó 0.25)
```

All metrics are normalized to a 0-100 scale for fair cross-language comparison.

## üîß Setup & Installation

### Required Libraries

````python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Display settings for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

print("‚úÖ All libraries imported successfully!")
print("‚úÖ Visualization settings configured!")
print("\nüì¶ Libraries loaded:")
print("   - pandas (data manipulation)")
print("   - numpy (numerical operations)")
print("   - matplotlib (static plots)")
print("   - seaborn (statistical visualizations)")
print("\nüé® Ready to analyze!")
````

### Load the Dataset

````python
# Load the combined dataset
data_file = '../data/raw/repositories_enriched.csv'

try:
    df = pd.read_csv(data_file)
    print(f"‚úÖ Dataset loaded successfully!")
    print(f"üìä Shape: {df.shape[0]} repositories √ó {df.shape[1]} columns")
    print(f"\nüîç Quick peek at the data:")
    print(f"   - Languages: {df['language'].nunique()}")
    print(f"   - Total Stars: {df['stars'].sum():,}")
    print(f"   - Total Forks: {df['forks'].sum():,}")
    print(f"   - Date Range: {df['created_at'].min()[:4]} to {df['created_at'].max()[:4]}")
    print(f"\nüìà Score Ranges:")
    print(f"   - Overall Score: {df['overall_score'].min():.2f} to {df['overall_score'].max():.2f}")
    print(f"   - Avg Overall Score: {df['overall_score'].mean():.2f}")
except FileNotFoundError:
    print("‚ùå Error: Data file not found!")
    print(f"   Please ensure '{data_file}' is in the data/raw directory.")
except Exception as e:
    print(f"‚ùå Error loading data: {e}")
````

---

## üèÜ Top Language Rankings Preview

Let's see which languages are leading the ecosystem!

````python
# Calculate language rankings by overall score
language_rankings = df.groupby('language').agg({
    'overall_score': 'mean',
    'popularity_score': 'mean',
    'activity_score': 'mean',
    'health_score': 'mean',
    'stars': 'sum',
    'forks': 'sum'
}).round(2)

# Sort by overall score
language_rankings = language_rankings.sort_values('overall_score', ascending=False)

# Display the rankings
print("üèÜ LANGUAGE RANKINGS BY OVERALL SCORE")
print("=" * 80)
print(f"\n{'Rank':<6} {'Language':<15} {'Overall':<10} {'Popularity':<12} {'Activity':<10} {'Health':<8}")
print("-" * 80)

for idx, (lang, row) in enumerate(language_rankings.iterrows(), 1):
    emoji = "ü•á" if idx == 1 else "ü•à" if idx == 2 else "ü•â" if idx == 3 else f"{idx:2d}."
    print(f"{emoji:<6} {lang:<15} {row['overall_score']:>8.2f}  {row['popularity_score']:>10.2f}  {row['activity_score']:>10.2f}  {row['health_score']:>7.2f}")

print("\n" + "=" * 80)
print("üéØ Key Insights:")
print(f"   ‚Ä¢ Top Language: {language_rankings.index[0]} ({language_rankings.iloc[0]['overall_score']:.2f})")
print(f"   ‚Ä¢ Highest Activity: {language_rankings['activity_score'].idxmax()} ({language_rankings['activity_score'].max():.2f})")
print(f"   ‚Ä¢ Most Popular: {language_rankings['popularity_score'].idxmax()} ({language_rankings['popularity_score'].max():.2f})")
print(f"   ‚Ä¢ Best Health: {language_rankings['health_score'].idxmax()} ({language_rankings['health_score'].max():.2f})")
````

### üìä Quick Visualization: Top Languages Comparison

````python
# Create a visualization of the top languages
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Overall Scores
ax1 = axes[0]
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(language_rankings)))
language_rankings['overall_score'].plot(kind='barh', ax=ax1, color=colors)
ax1.set_xlabel('Overall Score', fontsize=12, fontweight='bold')
ax1.set_ylabel('')
ax1.set_title('üèÜ Language Rankings by Overall Score', fontsize=14, fontweight='bold', pad=20)
ax1.grid(axis='x', alpha=0.3)
ax1.invert_yaxis()

# Add value labels
for idx, (lang, score) in enumerate(language_rankings['overall_score'].items()):
    ax1.text(score + 0.5, idx, f'{score:.1f}', va='center', fontsize=9, fontweight='bold')

# Plot 2: Score Components Comparison (Top 5)
ax2 = axes[1]
top5 = language_rankings.head(5)
x = np.arange(len(top5))
width = 0.25

ax2.bar(x - width, top5['popularity_score'], width, label='Popularity', color='#FF6B6B', alpha=0.8)
ax2.bar(x, top5['activity_score'], width, label='Activity', color='#4ECDC4', alpha=0.8)
ax2.bar(x + width, top5['health_score'], width, label='Health', color='#95E1D3', alpha=0.8)

ax2.set_xlabel('Programming Language', fontsize=12, fontweight='bold')
ax2.set_ylabel('Score', fontsize=12, fontweight='bold')
ax2.set_title('üìä Top 5 Languages: Score Components', fontsize=14, fontweight='bold', pad=20)
ax2.set_xticks(x)
ax2.set_xticklabels(top5.index, rotation=45, ha='right')
ax2.legend(loc='upper right', frameon=True, shadow=True)
ax2.grid(axis='y', alpha=0.3)
ax2.set_ylim(0, 100)

plt.tight_layout()
plt.show()

print("\n‚ú® Visualization complete!")
print("üí° For more detailed visualizations, explore the other notebooks!")
````

## üìå Key Findings Summary

Based on the analysis of 1,200 repositories across 12 languages:

````python
# Generate key insights
print("üéØ KEY FINDINGS FROM THE ANALYSIS")
print("=" * 80)

# Top performers
top_lang = language_rankings.index[0]
top_score = language_rankings.iloc[0]['overall_score']

print(f"\nü•á TOP PERFORMER: {top_lang}")
print(f"   ‚Ä¢ Overall Score: {top_score:.2f}")
print(f"   ‚Ä¢ Strength: {language_rankings.iloc[0][['popularity_score', 'activity_score', 'health_score']].idxmax().replace('_score', '').title()}")

# Most popular
most_pop = df.groupby('language')['stars'].sum().idxmax()
most_pop_stars = df.groupby('language')['stars'].sum().max()
print(f"\n‚≠ê MOST POPULAR: {most_pop}")
print(f"   ‚Ä¢ Total Stars: {most_pop_stars:,.0f}")
print(f"   ‚Ä¢ Average per repo: {most_pop_stars/100:,.0f}")

# Most active
most_active = language_rankings['activity_score'].idxmax()
activity_score = language_rankings.loc[most_active, 'activity_score']
print(f"\n‚ö° MOST ACTIVE: {most_active}")
print(f"   ‚Ä¢ Activity Score: {activity_score:.2f}")
print(f"   ‚Ä¢ Indicates: Strong ongoing development and community engagement")

# Best health
best_health = language_rankings['health_score'].idxmax()
health_score = language_rankings.loc[best_health, 'health_score']
print(f"\nüè• BEST REPOSITORY HEALTH: {best_health}")
print(f"   ‚Ä¢ Health Score: {health_score:.2f}")
print(f"   ‚Ä¢ Indicates: Strong documentation and community standards")

# Score spread
score_std = df['overall_score'].std()
print(f"\nüìä SCORE DISTRIBUTION:")
print(f"   ‚Ä¢ Mean: {df['overall_score'].mean():.2f}")
print(f"   ‚Ä¢ Std Dev: {score_std:.2f}")
print(f"   ‚Ä¢ Range: {df['overall_score'].min():.2f} - {df['overall_score'].max():.2f}")
print(f"   ‚Ä¢ Variance indicates significant differences between languages")

print("\n" + "=" * 80)
print("üí° For detailed analysis, proceed to the specialized notebooks!")
````
